```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
exceldata = pd.read_excel("./Datasets/data.xlsx")
```

```python
exceldata.head(10)
```

## Gender Analysis

```python
new_gender_header = exceldata.iloc[1]
gender_df = exceldata[4:6] 
gender_df.columns = new_gender_header
gender_df.columns.values[1] = 'Gender'
gender_df = gender_df.iloc[:, 1:10]
gender_df = gender_df.reset_index(drop=True)
gender_df
```

```python
columns_to_plot = ['All Wghtd %', 'PHEV Wghtd %', 'Tesla Wghtd %', 'Non-Tesla BEV Wghtd %']

# Create a figure and set of subplots
fig, axes = plt.subplots(nrows=1, ncols=len(columns_to_plot), figsize=(15, 5))

# Plot pie charts
for ax, column in zip(axes, columns_to_plot):
    ax.pie(gender_df[column], labels=gender_df['Gender'], autopct='%1.1f%%', startangle=140)
    ax.set_title(column)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

### Findings
- Majority of people buying Hybrid or Full electric vehicles are male.
- This parity is more prominant in case of Tesla buyers.

```python
new_age_header = exceldata.iloc[1]
age_df = exceldata[7:15] 
age_df.columns = new_age_header
age_df.columns.values[1] = 'Age Groups'
age_df = age_df.iloc[:, 1:10]
age_df = age_df.reset_index(drop=True)
age_df
```

```python
columns_to_plot = ['All Wghtd #', 'PHEV Wghtd #', 'Tesla Wghtd #', 'Non-Tesla BEV Wghtd #']

# Create a figure and set of subplots
fig, axes = plt.subplots(nrows=1, ncols=len(columns_to_plot), figsize=(15, 5), sharey=True)

# Plot bar plots
for ax, column in zip(axes, columns_to_plot):
    # Ensure the column data and x labels are strings for proper plotting
    ax.bar(age_df['Age Groups'].astype(str), age_df[column].astype(float))
    ax.set_title(column)
    ax.set_xlabel('Age Groups')
    ax.set_ylabel('Count')

    ax.set_xticks(age_df['Age Groups'])  # Set x-ticks to the Age Groups
    ax.set_xticklabels(age_df['Age Groups'], rotation=45, ha='right')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

### Findings 
- Majority of car buyers are between 30-70 years of age.
- Majority of Tesla buyers are middle age people compared to young and older. Younger people might not able to afford it and older people might be habitual to older technology.

## Income Analysis

```python
new_income_header = exceldata.iloc[1]
income_df = exceldata[22:31] 
income_df.columns = new_income_header
income_df.columns.values[1] = 'Income'
income_df = income_df.iloc[:, 1:10]
income_df = income_df.reset_index(drop=True)
columns_to_replace = ['PHEV Wghtd %',	'PHEV Wghtd #',	'Tesla Wghtd %',	'Tesla Wghtd #',	'Non-Tesla BEV Wghtd %',	'Non-Tesla BEV Wghtd #']
income_df[columns_to_replace] = income_df[columns_to_replace].replace({',': ''}, regex=True)
income_df
```

```python
columns_to_plot = ['PHEV Wghtd #', 'Tesla Wghtd #', 'Non-Tesla BEV Wghtd #']

# Create a figure and set of subplots
fig, axes = plt.subplots(nrows=1, ncols=len(columns_to_plot), figsize=(15, 5), sharey=True)

# Plot bar plots
for ax, column in zip(axes, columns_to_plot):
    # Ensure the column data and x labels are strings for proper plotting
    ax.bar(income_df['Income'].astype(str), income_df[column].astype(float))
    ax.set_title(column)
    ax.set_xlabel('Income')
    ax.set_ylabel('Count')

    ax.set_xticks(income_df['Income'])  # Set x-ticks to the Age Groups
    ax.set_xticklabels(income_df['Income'], rotation=45, ha='right')
    
# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

### Finding
- From above graphs income distribution for Hybrid Electric vehicle and Non Tesla BEV is approximately normal distribution.
- Majority of people buying Hybrid Electric vehicle and Non Tesla BEV have income between 100K - 150K.
- As income level increases people do not prefer Hybrid Electric vehicle and Non Tesla BEV.
- On other hand, most people with income higher than 100K prefer Tesla
